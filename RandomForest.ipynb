```python
#import all libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read the data file
df = pd.read_csv('creditcard.csv')
```

```python
df.head(4)
```

```python
df.describe()
```

```python
#check the data balance 
sns.countplot(x = 'Class', data = df, palette = 'magma')
plt.title('Fraud')
plt.show()
```

```python
#The data is highly imbalanced. We are going to use oversampling of minority class to balance it.
#This not required for random forest but doing it for practice
x = df.drop(columns=['Class'])
y = df['Class']

# Split data into train and test
from sklearn.model_selection import train_test_split
x_train, x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)

#creating synthetic samples
# concatenate our training data back together
X = pd.concat([x_train, y_train], axis=1)


# separate minority and majority classes
not_fraud = X[X['Class']==0]
fraud = X[X['Class']==1]

from sklearn.utils import resample
new_sample= resample(fraud, replace=True, # sample with replacement
                          n_samples=len(not_fraud), # match number in majority class
                          random_state=27) # reproducible results

final_train = pd.concat([not_fraud,new_sample],axis=0)
```

```python
x1_train = final_train.drop(labels = ["Class"],axis = 1)
y1_train = final_train['Class'].values

#scale the data
from sklearn.preprocessing import StandardScaler
sc_x= StandardScaler()
x1_train = sc_x.fit_transform(x1_train)
x_test = sc_x.transform(x_test)
```

```python
#prepare random forest
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=10).fit(x1_train,y1_train)

#predict
y_pred = model.predict(x_test)
```

```python
#prediction probabilities
y_prob = model.predict_proba(x_test)[:,1]

#check model performance 
from sklearn.metrics import roc_auc_score,accuracy_score,f1_score,recall_score

# Calculate roc auc. This is most useful for random forest. 
roc_value = roc_auc_score(y_test, y_prob)

#calculate accuracy score
acc = accuracy_score(y_test, y_pred)

#calculate f1_score
f1 = f1_score(y_test, y_pred)

#calculate recall score

recall = recall_score(y_test, y_pred)
```

```python
print('roc_score: ' +str(roc_value)+ ' ', 'accuracy_score: ' + str(acc) + ' ', 'f1_score: '+ str(f1) + ' ', 'recall_score: ' + str(recall)  )
```
